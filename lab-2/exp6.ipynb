# Experiment 6
Identify inconsistent or incorrect values in the dataset and correct them using the `replace()` function.

```python
import pandas as pd

df = pd.read_csv("data.csv")

df["Gender"] = df["Gender"].replace({
    "F": "Female",
    "M": "Male",
    "male": "Male",
    "female": "Female"
})

print("Corrected gender column:")
print(df["Gender"])
```
