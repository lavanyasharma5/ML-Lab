# Experiment 10
Create a boxplot for numerical attributes to identify potential outliers.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("data.csv")

# Ensure Salary is numeric and drop missing values
df["Salary"] = pd.to_numeric(df["Salary"], errors="coerce")
salary_data = df["Salary"].dropna()

# Create boxplot
plt.boxplot(salary_data)
plt.title("Boxplot of Salary")
plt.ylabel("Salary")
plt.show()
```
