# Experiment 1
Create a NumPy array using a Python list. Display the array and print its shape and data type.+++++

```python
import numpy as np

# Creating a NumPy array from a Python list
data = [10, 20, 30, 40, 50]
arr = np.array(data)

# Display array
print("Array:", arr)

# Display shape and data type
print("Shape:", arr.shape)
print("Data Type:", arr.dtype)
```
