# Experiment 5
Create a Pandas DataFrame using a dictionary containing student names and marks. Display the DataFrame and the data types of each column.

```python
import pandas as pd

# Dictionary data
data = {
    "Student Name": ["Aman", "Riya", "Karan", "Sneha"],
    "Marks": [85, 92, 78, 88]
}

# Create DataFrame
df = pd.DataFrame(data)

print("DataFrame:")
print(df)

print("\nData Types:")
print(df.dtypes)
```
