# Experiment 4
Generate a NumPy array of 10 random integers. Find the minimum, maximum, mean, and sum of the array.

```python
import numpy as np

# Create array of 10 random integers between 1 and 100
arr = np.random.randint(1, 101, 10)

print("Array:", arr)
print("Minimum:", arr.min())
print("Maximum:", arr.max())
print("Mean:", arr.mean())
print("Sum:", arr.sum())
```
